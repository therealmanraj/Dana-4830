```python
import boto3
import sagemaker
from sagemaker.sklearn.estimator import SKLearn

# Create a boto3 session in us-east-2 (the bucket's region)
boto_session = boto3.Session(region_name="us-east-2")
sagemaker_session = sagemaker.Session(boto_session=boto_session)

# Replace with your actual SageMaker execution role ARN.
role = "arn:aws:iam::222634404112:role/SageMakerExecutionRole"

# S3 path where your training data is stored.
data_path = "s3://dana-minicapstone/data/occupancy_model_zones.csv"

# Define a custom training job name.
job_name = "occupancy-training-job-20250403"

# Create an SKLearn estimator including additional dependencies.
sklearn_estimator = SKLearn(
    entry_point="occupancy_model.py",     # Your training script filename
    source_dir=".",                       # Upload everything in the current directory
    dependencies=["requirements.txt"],    # Ensure requirements.txt is in your local folder
    role=role,
    instance_type="ml.m5.xlarge",         # Choose an instance type based on your needs
    framework_version="0.23-1",           # Version of scikit-learn in the container
    output_path="s3://dana-minicapstone/model-artifacts/",
    sagemaker_session=sagemaker_session,
    job_name=job_name                     # Specify your custom training job name here
)

# Launch the training job.
sklearn_estimator.fit()
```
